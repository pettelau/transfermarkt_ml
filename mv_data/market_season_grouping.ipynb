```python
import pandas as pd

df_seasons= pd.read_csv('cleaned_player_season_data_grouped.csv')

df_mv= pd.read_csv('mv_data_all.csv')

df_mv.drop(columns='Club', inplace=True)

result_df = df_seasons.merge(df_mv, left_on=['Player_id', 'Season'],right_on=['Player_id', 'Season'], how='inner')

# save the result to a new csv file
result_df.to_csv('player_season_with_value.csv', index=False)
```
